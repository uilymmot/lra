```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from sklearn.manifold import TSNE
import copy
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import SparsePCA
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.decomposition import NMF
from sklearn.ensemble import RandomForestRegressor
import subprocess
from sklearn.model_selection import RandomizedSearchCV
import sys
import math
import scipy.spatial.distance
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.tree import DecisionTreeRegressor
import os
import shutil
import glob
import shap
from scipy.stats import pearsonr


cmap='viridis'
sys.path.insert(1, '../')
from LibHelperFuncs import *

from iterative_spectral_method.src import *
from iterative_spectral_method.sdr import *

rstate = 0
np.random.seed(0)
njob = 20
global_trees = 10

plt.rcParams['axes.grid'] = False
```

```python
originalDat = np.genfromtxt("../../Data/data_bas_preprocessing.csv", delimiter=',')
originalDat_pd = pd.read_csv("../../Data/data_bas_preprocessing_pd.csv")

X = originalDat[:,:-1]
Y = originalDat[:,-1]
C = originalDat_pd.columns

lower_thres = 1e-3

files = []
objectives = ['ISM_gaus_homogeneous_0.5.csv', 'pca_sparse_homogeneous_0.5.csv',\
              "ISM_squa_homogeneous_0.5.csv", "ISM_poly_homogeneous_0.5.csv",\
              'pca_standard.csv', 'nmf_H_matrix_homogeneous_0.5.csv', 'pca_homogeneous_0.5.csv']
```

```python
poly_k = kernels.polynomial.polynomial
gaus_k = kernels.gaussian.gaussian
squa_k = kernels.squared.squared
line_k = kernels.linear.linear
```

```python
fnumbers = range(1, 102, 10)
z = len(fnumbers)
```

```python
scores = []

for i in fnumbers:
    pca = PCA(n_components=i)
    nmf = NMF(n_components=i)
    
    X_pca = pca.fit_transform(X)
    X_nmf = nmf.fit_transform(X)

    sdr_class = sdr(X, Y, gaus_k, q = i)
    sdr_class.train()
    X_gaussian = sdr_class.get_reduced_dim_data(X)
    
    sdr_class = sdr(X, Y, squa_k, q = i)
    sdr_class.train()
    X_square = sdr_class.get_reduced_dim_data(X)
    
    sdr_class = sdr(X, Y, line_k, q = i)
    sdr_class.train()
    X_line = sdr_class.get_reduced_dim_data(X)
    
    sdr_class = sdr(X, Y, poly_k, q = i)
    sdr_class.train()
    X_poly = sdr_class.get_reduced_dim_data(X)
    
    print(i)
    
    scores.append(rf_model_crosseval(X_pca, Y, global_trees))
    scores.append(rf_model_crosseval(X_nmf, Y, global_trees))
    scores.append(rf_model_crosseval(X_gaussian, Y, global_trees))
    scores.append(rf_model_crosseval(X_square, Y, global_trees))
    scores.append(rf_model_crosseval(X_line, Y, global_trees))
    scores.append(rf_model_crosseval(X_poly, Y, global_trees))
```

```python
plt.figure(figsize=(14,10))

plt.plot(fnumbers, scores[0::6], label="PCA")
plt.plot(fnumbers, scores[1::6], label="NMF")
plt.plot(fnumbers, scores[2::6], label="ISM Gaussian")
plt.plot(fnumbers, scores[3::6], label="ISM Square")
plt.plot(fnumbers, scores[4::6], label="ISM Line")
plt.plot(fnumbers, scores[5::6], label="ISM Polynomial")

plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

plt.xlabel("Number of features", fontsize=24)
plt.ylabel("Cross Validated RMSE Score", fontsize=24)

plt.xlim([0,51])

plt.legend(fontsize=24)

plt.savefig("../../Writeups/Figures/DR_comparison_cv.pdf", bbox_inches='tight')
```
